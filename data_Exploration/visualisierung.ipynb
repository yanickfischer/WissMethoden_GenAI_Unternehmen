# Datenbereinigung & Vorbereitung

```python
# Erkennung aller Spalten mit ausschließlich 'Ja'/'Nein' als Werte
bool_cols = [
    col for col in df.columns
    if df[col].dropna().isin(['Ja', 'Nein']).all()
]

# Umwandlung in boolesche Werte
df[bool_cols] = df[bool_cols].applymap(lambda x: True if x == 'Ja' else False)

print(df)
```
